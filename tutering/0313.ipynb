```python
import pandas as pd
df = pd.DataFrame( {
    '이름': ['피카츄','라이츄','리자드','리자몽','꼬부기','어니부기'],
    '속성': ['전기', '전기', '불','불','물',None], 
    '색깔': ['노랑', '노랑', None,'빨강','파랑','파랑'],
    '순위': [ 5, 2, 6, 1, 4, 3] })

df.columns = ['이름', '속성','색깔','순위']
df
```

# Groupby
df.groupby('그룹화를 할 열의 이름')
- 데이터를 그룹화하여 그룹화에 대한 연산(max/min/mean/sum)을 수행한다.


```python
import pandas as pd
import numpy as np

data = {'학년': ['1학년', '1학년', '2학년', '2학년', '3학년', '3학년', '3학년'], 
        '성별': ['남', '여', '남', '여', '남', '여', '여'], 
        '영어 성적': [80, 90, 70, 60, 85, 95, 75]}
df = pd.DataFrame(data)
```

```python
# case1 학년을 기준으로 데이터를 그룹화하고 그룹화된 데이터에 대해 영어성적의 평균 도출
df.groupby('학년').mean('영어')
```

```python
# case2
df.groupby(df['학년'])['영어 성적'].mean()
```

```python
# case3
df[['영어 성적']].groupby(df['학년']).mean()
```

### Agg
-Agg는 사용자가 다중집계 작업을 간단하게 나타낼수있도록한다
-agg(적용하고싶은 함수)

```python
# 전체 학년의 영어성적의 합계/평균/최대값 도출
result = df.groupby('학년')['영어 성적'].agg(['sum', 'mean', 'max'])
result
```

```python
result = df.groupby('학년')['영어 성적'].agg(['sum', 'mean', 'max']).reset_index()
```

```python
# 예제 직원 데이터 생성
np.random.seed(0) # 일관된 결과를 위한 랜덤 시드 설정
departments = ['HR', 'Tech', 'Finance', 'Marketing']
genders = ['Male', 'Female']
# 100명의 직원 데이터 생성
data = {
    'EmployeeID': range(1, 101),
    'Department': np.random.choice(departments, 100),
    'Gender': np.random.choice(genders, 100),
    'Salary': np.random.randint(50000, 150000, 100) # 연봉 범위: 50,000 ~ 150,000
}

# 판다스 DataFrame으로 변환
employees_df = pd.DataFrame(data)

employees_df.head()
```

```python
# 부서별로 직원의 수를 구하세요
employees_df.groupby('Department')[['EmployeeID']].count()
```

```python
# 부서별로 직원의 연봉평균과 최대 연봉을 구하세요
employees_df.groupby('Department')['Salary'].agg(['mean', 'max'])
```

```python
# 부서와 성별을 기준으로 그룹화하여 각 그룹의 평균 연봉을 구하세요
employees_df.groupby(['Department', 'Gender'])[['Salary']].mean()
```

### concat vs merge
# concat
- pd.concat([df1,df2,...],axis=0(기본값)) -> 가로로 붙이고싶다면 axis=1
# merge
- pd.merge(df1,df2,on='공통컬럼',how='합치는방법')

```python
a=[100,55,80,97]
b=[26,53,45,100]
c=[94,100,32,43]

df1=pd.DataFrame([a,b,c],columns=['korean','math','english','history'])
df1
```

```python
a=[88,94,21,38]
b=[26,53,44,50]
c=[90,30,32,45]

df2=pd.DataFrame([a,b,c],columns=['korean','math','english','history'])
df2
```

```python
a=[88,45,67,38]
b=[26,34,46,56]
c=[30,30,36,99]

df3=pd.DataFrame([a,b,c],columns=['korean','math','english','history'])
df3
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3]).reset_index(drop=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
# df1의 english를 기준으로 df2와 붙여보자 how안써도됨
pd.merge(df1, df2, on='english', how='left')
```
