### datatime에서 사용되는 format
- %Y : 년도를 전체표시 (2024)
- %y : 년도를 축약해서 표시 (24)
- %b : 축약된 월의 이름(Mar)
- %B : 월의 이름(March)
- %m : 월을 숫자로 표시 (02)
- %d : 일을 숫자로 표시 (19)
- %H : 24시를 기준으로 시를 표시
- %I : 12시를 기준으로 시를 표시
- %p : AM/PM을 표시
- %M : 분을 표시
- %S(대문자): 초를 표시
- %a : 요일의 이름을 축약해서 표시(Mon)
- %A : 요일의 이름을 표시 (Monday)
- %w : 요일을 숫자로 표시(0: 일요일, ..., 6: 토요일)

### 연습문제
1. csv 폴더에있는 Sales Records.csv 파일 로드
2. 'Order ID','Total Revenue','Total Cost','Total Profit','Order Priority' 컬럼을 제거
3. 컬럼의 이름을 변경 -> ['대륙','국가','상품종류','판매채널','주문일자','발송일자','판매개수','단가','원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 기존의 인덱스는 제거하고 인덱스를 재지정
6. 파생변수 생성
    - 총단가 = 판매개수 * 단가
    - 총원가 = 판매개수 * 원가
    - 총이윤 = 총단가 - 총원가

```python
import pandas as pd
```

```python
df=pd.read_csv("../csv/Sales Records.csv")
```

```python
df=df.drop(columns=['Order ID','Total Revenue','Total Cost','Total Profit','Order Priority'], axis=1)
```

```python
df.columns=['대륙','국가','상품종류','판매채널','주문일자','발송일자','판매개수','단가','원가']
```

```python
df
```

```python
df.sort_values(['대륙','국가'], ascending=True, inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df['총단가']=(df['판매개수']*df['단가']).fillna(0)
df['총원가']=(df['판매개수']*df['원가']).fillna(0)
df['총이윤']=(df['총단가']-df['총원가']).fillna(0)
```

```python
df
```

```python
# 데이터프레임의 그룹화
# groupby({그룹화할 컬럼의 이름}).연산식함수
#df.loc[:['대륙','총이윤']]
#df.iloc[:,[0, -1]]
group_data=df[['대륙','총이윤']].groupby('대륙').sum()
```

```python
group_data.sort_values('총이윤',ascending=False)
```

```python
# 한국의 총이윤 순위 찾기
group_data2=df[['국가','총이윤']].groupby('국가').sum()
```

```python
group_data2.sort_values('총이윤',ascending=False, inplace=True)
```

```python
group_data2.index
```

```python
# case1 for문을 이용하여 랭크를 찾아본다
korea_rank=1

for i in group_data2.index:
    # i가 의미하는 것은? 국가명
    if i == "South Korea":
        break
    korea_rank += 1
print(korea_rank)
```

```python
# case2 reset_index() 함수를 이용하여 랭크
case2=group_data2.reset_index()
```

```python
flag=case2['국가']=='South Korea'
case2.loc[ flag, ].index[0]+1
```

```python
# case3 파생변수를 생성
group_data2['rank']=range(1, len(group_data2)+1)
```

```python
group_data2.loc['South Korea', 'rank']
```

```python
# case4 list 안에 내장함수를 이용
list(group_data2.index).index("South Korea")+1
```

```python
# 대륙, 국가로 그룹화
group_data3=df[['대륙','국가','총이윤']].groupby(['대륙','국가']).sum()
```

```python
group_data3.unstack()
```

```python
group_data3.stack()
```

```python
# pivot_table()
# pandas 안에 내장되어있는 함수

group_data4=pd.pivot_table(df, index=['대륙','국가'], columns='상품종류', values='총이윤', aggfunc=["sum","mean"], margins=True) # margins는 맨 끝에 총합계 표시
```

```python
# groupby에서 연산식이 2개 이상인 경우
group_data5=df[['대륙','국가','상품종류','총이윤']].groupby(['대륙','국가','상품종류']).agg(['sum', 'mean'])
```
