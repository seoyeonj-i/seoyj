```python
# 라이브러리 로드
import matplotlib.pyplot as plt # 해당하는 라이브러리에서 기능 하나(pyplot)만 사용할 때
# from matplotlib import pyplot # 해당하는 라이브러리에서 기능 여러개 쓸 때
```

```python
# 라인 그래프
# plot({1차원 데이터형태})
# plot의 1차원 데이터 인자의 개수가 1개인 경우에는 y축 데이터로 대입, x축은 0부터 1씩 증가하는 데이터 생성
# 인자의 개수가 2개인 경우에는 첫번째 인자가 x축 데이터, 두번째 인자가 y축 데이터가 된다

data=[5,3,7,1]

plt.plot(data)
```

```python
# 2개의 그래프를 출력
plt.plot(data)
plt.plot([10,2,4,7])
plt.show()
```

```python
plt.plot([1,3,5,7],data)
plt.show()
```

```python
# 라인 그래프에서 라인의 색을 변경
plt.plot(data, "red")
plt.show()
```

```python
# 선의 종류를 변경
plt.plot(data, "--r")
plt.show()
```

```python
# x,y에 라벨 설정
# 그래프의 제목 설정
plt.plot(data)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gragh')
plt.show()
```

```python
# 막대 그래프 # x,y 중 하나의 값만 넣었을때 에러발생 => 두개다 넣어야함
# bar({x축데이터}, {y축데이터})
x=[1,2,3,4]
y=[80,90,70,80]

plt.bar(x,y)
plt.show()
```

```python
# 막대의 색상 변경
plt.bar(x,y, color='red') # 여러개의 색상 color=['red','yellow','green','blue']
plt.show

# 막대의 너비 변경
plt.bar(x,y, width=0.4)
plt.show()
```

```python
plt.bar(['A','B','C','D'], y)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
# csv 폴더 안에 있는 drinks.csv 로드
drinks=pd.read_csv("../csv/drinks.csv")
```

```python
drinks.info()
```

```python
# 데이터에서 상관계수를 출력
# 데이터프레임 안에 내장된 corr() 함수 사용
corr=drinks.iloc[:, 1:5].corr(method='pearson')
corr
```

```python
# 히트맵 출력
cols_view=['bear','spirit','wine','alcohol']

hm=sns.heatmap(corr.values, cbar=True,annot=True,square=True, fmt='.3f', annot_kws={'size':20}, xticklabels=cols_view, yticklabels=cols_view) 
#cbar 사이드바 유무 # annot 히트맵에 숫자표시 유무 #square 히트맵을 정사각형으로 표시 유무 # fmt 숫자의 소수점 자리수 # annot 숫자의 크기
plt.tight_layout()
plt.show()
```

```python
# 데이터의 분포도를 시각화
sns.pairplot(drinks.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol'], height=2.5,corner=True,kind='reg',plot_kws={'line_kws':{'color':'red'}})
plt.show()
```
