# 일일확진자, 일일사망자를 생성하는 함수
- 매개변수 2개(데이터프레임, 오름차순정렬기준컬럼)
- 특정컬럼을 제거
    -unnamed:0 컬럼이 존재한다면제거, 존재하지않으면 아무행동도 하지 않는다
    - seq 컬럼이 존재한다면 제거, 존재하지 않으면 아무행동도 하지않는다
- 컬럼의 이름을 변경
- 2번째 매개변수를 기준으로 데이터프레임을 오름차순으로 정렬
- 인덱스를 초기화하고 기존의 인덱스는 제거
- '일일사망자','일일확진자' 컬럼을 추가
- 모든 작업이 끝난 데이터 프레임을 리턴

```python
def corona_EDA(df,_col):
    _df=df.copy()
    # 특정 컬럼을 제거
    # unnamed: 0 컬럼이 존재하는가?
    if "Unnamed: 0" in _df.columns:
        _df.drop('Unnamed: 0', axis=1, inplace=True)
# seq 컬럼이 존재하는가
    if "seq" in df.columns:
        _df.drop('seq',axis=1,inplace=True)
# 컬럼의 이름을 변경
    _df.columns=['등록일시','총사망자','총확진자','기준일','기준시간','수정일시','누적의심자','누적확진율']
# 2번째 매개변수를 기준으로 데이터 프레임을 오름차순으로 정렬
    _df.sort_values(_col, inplace=True)
# 인덱스를 초기화하고 기존의 인덱스는 삭제
    _df.reset_index(drop=True, inplace=True)
# 파생변수 생성
    _df['일일사망자']=_df['총사망자'].diff().fillna(0)
    _df['일일확진자']=_df['총확진자'].diff().fillna(0)
    return _df
```
